```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = datasets.load_iris()
```

```python
dataset
```

```python
cols = dataset['feature_names']
data = dataset['data']
target = dataset['target']
target_names = dataset['target_names']
```

```python
df = pd.DataFrame(data, columns=cols)
```

```python
display(df.head(3))
```

```python
pd.plotting.scatter_matrix(df,figsize=[10,10], c=target)
plt.show()
```

```python
df2 = df
df2['species'] = [target_names[v] for v in target]
```

```python
sns.set(style="ticks")

sns.pairplot(df2, size=2, hue='species')
plt.show()
```

```python
gnb = GaussianNB()
gnb.fit(data, target)
```

```python
p = gnb.predict(data)
```

```python
acc = (p == target).sum()/len(p)*100
print('Gaussian Naive Bayes is {0}% accurate.'.format(acc))
```
