```python
%config InlineBackend.figure_format = 'retina'

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
```

```python
# Import datasets to explore
df_2017_12 = pd.read_excel('../data/raw/EmployeePositionRoster_12312017.xls', header=0)
df_2016_12 = pd.read_excel('../data/raw/EmployeePositionRoster_12312016.xls', header=0)
df_2015_12 = pd.read_excel('../data/raw/EmployeePositionRoster_12312015.xls', header=0)
```

```python
# Make 2015 data have consistent naming
df_2015_12['ClsIndc'] = np.nan
col_renames = [name for name in df_2016_12.columns if name!='ClsIndc'] + ['ClsIndc']
df_2015_12.columns = col_renames
df_2015_12 = df_2015_12[df_2016_12.columns]
```

```python
# Add year to DF
df_2017_12['year'] = 2017
df_2016_12['year'] = 2016
df_2015_12['year'] = 2015
```

```python
display(df_2017_12)
display(df_2016_12)
display(df_2015_12)
```

```python
df = pd.concat([df_2017_12, df_2016_12, df_2015_12])

# Check that rows were all appended
assert(df.shape[0] == df_2017_12.shape[0] + df_2016_12.shape[0] + df_2015_12.shape[0])
```

```python
display(df.groupby(['Dept ID', 'Department', 'Job Title', 'year'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum','mean']).head())
```

```python
# format df and rename columns to prepare for pivot
temp = df.groupby(['Dept ID', 'Department', 'Job Title', 'year'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum','mean']).reset_index()
temp.columns = [' '.join(col).strip().lower().replace(' ', '_') for col in temp.columns.values]
# display(temp.head())

# create a pivot table:
pivot = pd.pivot_table(temp,
               index=['dept_id', 'department'],
               columns='year',
               values='annual_salary_sum',
               aggfunc='sum'
              )
display(pivot.head())

pivot['2016_delta'] = pivot[2016] - pivot[2015]
pivot['2017_delta'] = pivot[2017] - pivot[2016]


df_funding = pivot[['2016_delta','2017_delta']].reset_index()
```

```python
# Visualize change in funding in 2016 and 2017
pivot[['2017_delta', '2016_delta']].hist(bins=50, xrot=45, figsize=(15,5))
plt.xlim((-0.2e7, +0.2e7))
plt.show()
```

```python
# format df and rename columns to prepare for pivot
temp = df.groupby(['Dept ID', 'Department', 'Job Title', 'year'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum','mean']).reset_index()
temp.columns = [' '.join(col).strip().lower().replace(' ', '_') for col in temp.columns.values]
# display(temp.head())

# create a pivot table:
pivot = pd.pivot_table(temp,
               index=['dept_id', 'department'],
               columns='year',
               values='annual_salary_count',
               aggfunc='sum'
              )
display(pivot.head())

pivot['2016_delta_emps'] = pivot[2016] - pivot[2015]
pivot['2017_delta_emps'] = pivot[2017] - pivot[2016]

df_emps = pivot[['2016_delta_emps','2017_delta_emps']].reset_index()
```

```python
# Visualize change in employees in 2016 and 2017
pivot[['2017_delta_emps', '2016_delta_emps']].hist(bins=50, xrot=45, figsize=(15,5))
plt.xlim((-50, +50))
plt.show()
```

```python
df_deltas = pd.concat([df_funding, df_emps], axis=1)

print('Data shows that funding and personnel are linked - when funding decreases personnel does too')
print('This is critical, because our independent variables may be linked. This needs to be taken into account in any statistical modeling or the model could encounter numerical problems.')

plt.figure(figsize=[12,10])
ax1 = plt.subplot(2,2,1)
df_deltas.plot.scatter(x='2016_delta_emps', y='2016_delta', ax=ax1)
plt.title('delta salary vs. delta employees - 2016')
ax2 = plt.subplot(2,2,2)
df_deltas.plot.scatter(x='2017_delta_emps', y='2017_delta', ax=ax2)
plt.title('delta salary vs. delta employees - 2017')

ax1 = plt.subplot(2,2,3)
df_deltas.plot.scatter(x='2016_delta_emps', y='2016_delta', ax=ax1)
plt.title('delta salary vs. delta employees - 2016 zoomed in')
plt.xlim([-100,100])
plt.ylim([-0.5e7, 0.5e7])
ax2 = plt.subplot(2,2,4)
df_deltas.plot.scatter(x='2017_delta_emps', y='2017_delta', ax=ax2)
plt.title('delta salary vs. delta employees - 2017 zoomed in')
plt.xlim([-100,100])
plt.ylim([-0.5e7, 0.5e7])
plt.show()
```
