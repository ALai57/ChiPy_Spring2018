```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import datasets to explore
df_2017_12 = pd.read_excel('EmployeePositionRoster_12312017.xls', header=0)
df_2016_12 = pd.read_excel('EmployeePositionRoster_12312016.xls', header=0)
df_2015_12 = pd.read_excel('EmployeePositionRoster_12312015.xls', header=0)
```

```python
# Examine datasets
print(df_2017_12.head())
print(df_2016_12.head())
print(df_2015_12.head())
```

```python
df_2017_12.groupby(['Dept ID', 'Department', 'Job Title'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','mean'])
```

```python
df_2016_12.groupby(['Dept ID', 'Department', 'Job Title'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','mean'])
```

```python
df_2015_12.groupby(['UNIT NUMBER', 'UNIT NAME', 'JOB DESCRIPTION'])['ANNUAL SALARY', 'ANNUAL BENEFIT COST'].agg(['count','mean'])
```

```python
# Look at results by department
df_2017_sum = df_2017_12.groupby(['Dept ID', 'Department'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum']).reset_index()
df_2016_sum = df_2016_12.groupby(['Dept ID', 'Department'])['Annual Salary', 'Annual Benefit Cost'].agg(['count','sum']).reset_index()
df_2015_sum = df_2015_12.groupby(['UNIT NUMBER', 'UNIT NAME'])['ANNUAL SALARY', 'ANNUAL BENEFIT COST'].agg(['count','sum']).reset_index()
```

```python
df_2017_sum.columns = ['Dept ID', 'Department', '2017 n_emps','2017 salary','d','2017 benefits']
df_2016_sum.columns = ['Dept ID', 'Department','2016 n_emps','2016 salary','d','2016 benefits']
df_2015_sum.columns = ['Dept ID', 'Department','2015 n_emps','2015 salary','d','2015 benefits']

df_2017_sum.drop('d', axis=1, inplace=True)
df_2016_sum.drop('d', axis=1, inplace=True)
df_2015_sum.drop('d', axis=1, inplace=True)
```

```python
df_full = df_2017_sum.merge(df_2016_sum, on='Dept ID', how='outer').merge(df_2015_sum, on='Dept ID', how='outer')
df_full.drop(['Department_y','Department'], axis=1, inplace= True)

df_full['2017 total'] = df_full['2017 salary'] + df_full['2017 benefits']
df_full['2016 total'] = df_full['2016 salary'] + df_full['2016 benefits']
df_full['2015 total'] = df_full['2015 salary'] + df_full['2015 benefits']

df_full['2017 delta'] = df_full['2017 total'] - df_full['2016 total']
df_full['2016 delta'] = df_full['2016 total'] - df_full['2015 total']


df_full['2017 delta emps'] = df_full['2017 n_emps'] - df_full['2016 n_emps']
df_full['2016 delta emps'] = df_full['2016 n_emps'] - df_full['2015 n_emps']
```

```python
print(df_full['2015 total'].sum())
print(df_full['2016 total'].sum())
print(df_full['2017 total'].sum())
```

```python
growth_15_16 = df_full['2016 total'] > df_full['2015 total']
growth_16_17 = df_full['2017 total'] > df_full['2016 total']
```

```python
print('Data shows that there are many schools that received funding cuts and many that did not')
print('This is critical, because it shows that our data approximately evenly samples our sample space - we have many examples to learn from on both sides of the funding cut spectrum.')

plt.figure(figsize=[12,8])
ax1 = plt.subplot(2,2,1)
df_full['2017 delta'].plot.hist(ax=ax1, bins=50)
plt.title('Change in funding: 2017-2016')
plt.ylabel('# schools')
ax2 = plt.subplot(2,2,2)
df_full['2016 delta'].plot.hist(ax=ax2, bins=50)
plt.title('Change in funding: 2016-2015')
plt.ylabel('# schools')


ax1 = plt.subplot(2,2,3)
df_full['2017 delta'].plot.hist(ax=ax1, bins=np.arange(-25,25)*1e5)
plt.title('Change in funding: 2017-2016. Zoomed')
plt.ylabel('# schools')
plt.xlabel('$')
plt.xlim([-0.5e7, +0.5e7])
ax2 = plt.subplot(2,2,4)
df_full['2016 delta'].plot.hist(ax=ax2, bins=np.arange(-25,25)*1e5)
plt.title('Change in funding: 2016-2015. Zoomed')
plt.ylabel('# schools')
plt.xlabel('$')
plt.xlim([-0.5e7, +0.5e7])
plt.show()
```

```python
print('Data shows that there are many schools that received personnel cuts and many that did not')
print('This is critical, because it shows that our data approximately evenly samples our sample space - we have many examples to learn from on both sides of the personnel cut spectrum.')

plt.figure(figsize=[12,8])
ax1 = plt.subplot(2,2,1)
df_full['2017 delta emps'].plot.hist(ax=ax1, bins=50)
plt.title('Change in staff: 2017-2016')
plt.ylabel('# schools')
ax2 = plt.subplot(2,2,2)
df_full['2016 delta emps'].plot.hist(ax=ax2, bins=50)
plt.title('Change in staff: 2016-2015')
plt.ylabel('# schools')


ax1 = plt.subplot(2,2,3)
df_full['2017 delta emps'].plot.hist(ax=ax1, bins=np.arange(-25,25))
plt.title('Change in staff: 2017-2016. Zoomed')
plt.ylabel('# schools')
plt.xlabel('$')
plt.xlim([-20, +20])
ax2 = plt.subplot(2,2,4)
df_full['2016 delta emps'].plot.hist(ax=ax2, bins=np.arange(-25,25))
plt.title('Change in staff: 2016-2015. Zoomed')
plt.ylabel('# schools')
plt.xlabel('$')
plt.xlim([-20, +20])
plt.show()
```

```python
print('Data shows that funding and personnel are linked - when funding decreases personnel does too')
print('This is critical, because our independent variables may be linked. This needs to be taken into account in any statistical modeling or the model could encounter numerical problems.')

plt.figure(figsize=[12,10])
ax1 = plt.subplot(2,2,1)
df_full.plot.scatter(x='2016 delta emps', y='2016 delta', ax=ax1)
plt.title('delta salary vs. delta employees - 2016')
ax2 = plt.subplot(2,2,2)
df_full.plot.scatter(x='2017 delta emps', y='2017 delta', ax=ax2)
plt.title('delta salary vs. delta employees - 2017')

ax1 = plt.subplot(2,2,3)
df_full.plot.scatter(x='2016 delta emps', y='2016 delta', ax=ax1)
plt.title('delta salary vs. delta employees - 2016 zoomed in')
plt.xlim([-100,100])
plt.ylim([-0.5e7, 0.5e7])
ax2 = plt.subplot(2,2,4)
df_full.plot.scatter(x='2017 delta emps', y='2017 delta', ax=ax2)
plt.title('delta salary vs. delta employees - 2017 zoomed in')
plt.xlim([-100,100])
plt.ylim([-0.5e7, 0.5e7])
plt.show()
```
